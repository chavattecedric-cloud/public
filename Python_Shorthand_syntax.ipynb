Une fonction qui renvoie la somme des nombres pairs présents dans une liste contenant des nombres entiers.

```python
def nombre_pair (liste):
  total=0
  for i in liste:
    if i%2 ==0:
      total=total+i
  return total

nombre_pair([0, 3, -6, 9, 12, -15, 18, -21, 24, 27])
```

```python
resultat = lambda liste : sum([i for i in liste if i%2==0])
```

```python
(lambda liste: sum([i for i in liste if i%2==0]))([0, 3, -6, 9, 12, -15, 18, -21, 24, 27])
```

Une fonction qui convertit une liste de nombres entiers positifs ou négatifs) en une nouvelle liste contenant leurs valeurs absolues.

```python
def absolu (liste):
  nouvelle_liste=[]
  for i in liste:
    if i < 0:
      nouvelle_liste.append(abs(i))
    else:
      nouvelle_liste.append(i)
  return nouvelle_liste

absolu ([0, 3, -6, 9, 12, -15, 18, -21, 24, 27])
```

```python
nouvelle_liste = lambda liste :[abs(i) for i in liste]
```

```python
(lambda liste: [abs(i) for i in liste])([0, 3, -6, 9, 12, -15, 18, -21, 24, 27])
```
